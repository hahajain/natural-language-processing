# Tf-Idf example

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# using default tokenizer in TfidfVectorizer
#min_df - minimum document frequency
tfidf = TfidfVectorizer(min_df=2, max_df=0.5, ngram_range=(1, 2))
```

```python
texts = [
    "good movie", "not a good movie", "did not like", 
    "i like it", "good one", "bad one", "waste of money", "worth my money"
]

features = tfidf.fit_transform(texts)
pd.DataFrame(
    features.todense(),
    columns=tfidf.get_feature_names()
)
```

```python
texts = [
    "good movie", "not a good movie", "did not like", 
    "i like it", "good one", "bad one", "waste of money", "worth my money", "waste of time"
]

features = tfidf.fit_transform(texts)
pd.DataFrame(
    features.todense(),
    columns=tfidf.get_feature_names()
)
```

#Observe how adding "waste of time", brings in the waste of "waste of money" in picture, which was being ignored previously.
